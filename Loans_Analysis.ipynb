# Loan Dataset Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
import numpy as np
```

```python
analysis_df = pd.read_csv('../Loan-Dataset-Project/Data/Loans_for_Analysis.csv')

analysis_df = analysis_df.drop('Unnamed: 0', axis = 1)
```

# Loan Status vs Average Annual Income

## Though there is not a large difference, the average annual income for fully paid loans is a  it higher than those that were charged off

```python
sns.barplot(data=analysis_df, x='Loan Status', y='Annual Income', estimator = np.mean)
```

# Average Loan Amount per Loan Status

## The average loan amount for charged off loans is almost 25% higher than those that were fully paid

```python
sns.barplot(data=analysis_df, x='Loan Status', y='Current Loan Amount', estimator = np.mean)
```
